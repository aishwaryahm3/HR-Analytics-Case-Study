### Conclusion:

Employees may leave the company because they are less satisfied, haven't been promoted in the last five years, and other employees have experienced work accidents. The negative correlation between the `left` column and `satisfactoryLevel`, `workAccident`, and `promotionInLast5years` indicates that those who leave tend to have lower satisfaction, fewer promotions, and fewer work accidents.

## All in one

```python
chosen = ["dept","salary","promotionInLast5years","workAccident","timeSpent.company","numberOfProjects"]
```

```python
for i,j in enumerate(chosen):
    print(i,j)
```

```python
#creating subplots

fig, axes = plt.subplots(3, 2, figsize=(15, 15))

for p, q in enumerate(chosen):
    ax = axes[p // 2, p % 2]
    sns.countplot(x=q, data=df, hue='left', ax=ax)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.subplots_adjust(hspace=0.7)
plt.savefig('All_Subplots.png')
```

## Understanding:

Employees with less projects as well as employees with more number of projects are more likely to leave the company.

Years of exp: 3-6 years has to to be given more consideration.

Promotion -> Significant Role

Salary -> Low, Medium Salaried employees are much more likely to leave.
